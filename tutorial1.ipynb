```python
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt

img = np.array(Image.open('./Desktop/images/1.jpg'))
data = Image.fromarray(img) #convert nparray to image

img_red = img[:,:,0] # this can add constant value 
img_green = img[:,:,1]
img_blue = img[:,:,2]

data_red =Image.fromarray(img_red.astype('uint8'))
data_green = Image.fromarray(img_green.astype('uint8'))
data_blue = Image.fromarray(img_blue.astype('uint8'))

fig,axs = plt.subplots(1, 4, figsize=(20,5)) # row4 columns

#originalmage


axs[0].imshow(img)
axs[0].axis('off')
axs[0].set_title('originalimage')

axs[1].imshow(data_red, cmap='gray')
axs[1].axis('off')
axs[1].set_title('red channel')

axs[2].imshow(data_green, cmap='gray')
axs[2].axis('off')
axs[2].set_title('green channel')

axs[3].imshow(data_blue, cmap='gray')
axs[3].axis('off')
axs[3].set_title('blue channel')

plt.tight_layout()
plt.show()

```

```python
#create RGB image grom color channels
create_rgb = np.zeros((len(img_red),len(img_red[0]),3),'uint8')
create_rgb[:,:,0] = img_red
create_rgb[:,:,1] = img_green
create_rgb[:,:,2] = img_blue
img_created = Image.fromarray(create_rgb)

plt.imshow(img_created)
plt.axis('off')
plt.show()
```
